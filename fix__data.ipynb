```python
def j_fname(row):
    row = row.to_list()
    return row[0]+'-'+row[1].replace('.mp4', '.mp3')
    

fnames = vid_label.loc[:, 'video':'utterance'].apply(j_fname, axis=1)

a = pd.concat([vid_label.loc[:, 'arousal':], fnames, transcript['transcript']], axis=1)
a.columns = ['arousal','valence', 'EmotionMaxVote', 'video', 'transcript']
a
```
